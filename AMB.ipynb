```python
def update_trust(trust, player_id, last_p1_strategy, last_p2_strategy, last_p3_strategy):
    if player_id == 0:
        other_players_collaborated = last_p2_strategy == 1 and last_p3_strategy == 1
    elif player_id == 1:
        other_players_collaborated = last_p1_strategy == 1 and last_p3_strategy == 1
    else:
        other_players_collaborated = last_p1_strategy == 1 and last_p2_strategy == 1
    
    # Increase trust if others collaborated
    if other_players_collaborated:
        trust = min(1, trust + 0.1)
    # Decrease trust if others defected
    else:
        trust = max(0, trust - 0.1)
    
    return trust
```
